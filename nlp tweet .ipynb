```python
df=df.drop_duplicates()
df
```

```python
df['Tweet']=df['Tweet'].str.lower()
df[cl[0]]=df[cl[0]].str.lower()
df
```

```python
def conv(a):
    return re.sub('[^A-Za-z0-9]',' ',a)
```

```python
df['Tweet']=df['Tweet'].apply(conv)
```

```python
def conv1(a):
    ls=[]
    s=a.split()
    for i in s:
        if i not in stopwords.words('english'):
            ls.append(i)
    return ' '.join(ls)
```

```python
df['Tweet']=df['Tweet'].apply(conv1)
```

```python
X = df.iloc[:,1:3]
y = df['Category']
X
```

```python

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y = encoder.fit_transform(y)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
y_train
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
df1_tr = pd.DataFrame(data =tfidf.fit_transform(X_train['Tweet']).toarray())
df1_ts = pd.DataFrame(data =tfidf.transform(X_test['Tweet']).toarray())
df1_tr
```

```python
X_train=pd.concat((X_train.reset_index().drop('index',axis=1).iloc[:,:1],df1_tr),axis=1)
X_test=pd.concat((X_test.reset_index().drop('index',axis=1).iloc[:,:1],df1_ts),axis=1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```
